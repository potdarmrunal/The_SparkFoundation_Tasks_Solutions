```python
#Import Required Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#Read Match Dataset
match = pd.read_csv(r'F:\board Infinity\Excel Study\Dashboards\New folder\matches.csv')
match.head()
```

```python
#Read Deliveries Dataset

deliveries = pd.read_csv(r'F:\board Infinity\Excel Study\Dashboards\New folder\deliveries.csv')
match.head()
```

```python
# Create a new DataFrame called SeasonsData containing column id,season,winner
seasonsData  = match[['id','season','winner']]
```

```python
# Count Season Wise winning Team
count_of_winning_team_per_season = match.groupby('season')['winner'].value_counts()
count_of_winning_team_per_season
```

```python
# Plot Number Of Matches Played Per Season
plt.figure(figsize = (15,7))
sns.countplot('season',palette="viridis",data=match)
plt.xlabel("Seasons",fontsize=20)
plt.ylabel("Matches",fontsize=20)
plt.title("Number Of Matches Played Per Season",fontsize=20)
```

```python
#Number of matches won by team
plt.figure(figsize=(15,7))
sns.countplot('winner',data=match)
plt.xlabel('Teams',fontsize=18)
plt.ylabel('Count Of Winning Team',fontsize=18)
plt.xticks(rotation=50)
plt.title("Number Of Matches Win By Teams",fontsize=18)
```

```python
#Top 5  player of IPL match from 2008 to 2017
data = match.player_of_match.value_counts()[:5]
plt.figure(figsize=(15,7))
fig, ax = plt.subplots()
ax.set_ylim([0,20])
ax.set_ylabel("Count")
ax.set_title("Top 5 player of the match Winners")
data.plot.bar()
sns.barplot(x = data.index, y = data, orient='v', palette="hsv");
plt.show()
```

```python
#Create Dataframe final_matche containg "Final Matches" data

final_matches = match.drop_duplicates(subset=['season'], keep='last')
final_matches.head()
```

```python
# Percentage Of Final Match Winned From Toss 
Toss=final_matches.toss_winner.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
plt.figure(figsize = (7,6))
plt.pie(sizes, labels=labels,autopct='%1.1f%%')
plt.title('Toss Result',fontsize=20)
plt.axis('equal')
plt.show()
```

```python
# Percentage Of Final Match Winned From Field Selection
Toss=final_matches.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
plt.figure(figsize = (7,6))
plt.pie(sizes, labels=labels,autopct='%1.1f%%')
plt.title('Wiiner By Field Selection',fontsize=20)
plt.axis('equal')
plt.show()
```

```python
# Dismissals in IPL
plt.figure(figsize=(18,10))
ax=sns.countplot(deliveries.dismissal_kind)
plt.title("Dismissals in IPL",fontsize=20)
plt.xlabel("Dismissals kind",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
All_Data = deliveries.merge(seasonsData, left_on="match_id", right_on="id", how="inner")
All_Data.head()
```

```python

# we will print numbers of fours hit by team
four_data = All_Data[All_Data['batsman_runs']==4]
four_data.groupby('winner')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')]).sort_values(by="runs by fours",ascending=False)
```

```python
# we will plot graph of four hit by players

batsman_four = four_data.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
ax = batsman_four.iloc[:10,:]
ax.plot('batsman','four',kind='bar')
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=15)
plt.ylabel("No of four",fontsize=15)
plt.show()
```

```python
# we will print numbers of six hit by team
six_data = All_Data[All_Data['batsman_runs']==6]
six_data.groupby('winner')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')]).sort_values(by="runs by fours",ascending=False)
```

```python
# We will print the top 10 leading run scorer in IPL
batsman_score=All_Data.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum',ascending=False).reset_index(drop=True)
batsman_score=batsman_score.rename(columns={'sum':'batsman_runs'})
print("*** Top 10 Leading Run Scorer in IPL ***")
batsman_score.iloc[:10,:]
```

```python
# we will print ipl most wicket taking bowlers
wicket_data = All_Data.dropna(subset=['dismissal_kind'])
wicket_data = wicket_data[~wicket_data['dismissal_kind'].isin(['run out','retired hurt','obstructing the field'])]
wicket_data.groupby('bowler')['dismissal_kind'].agg(['count']).reset_index().sort_values('count',ascending=False).reset_index(drop=True).iloc[:10,:]
```

```python
# we will print no of matches played by batsman
No_Matches_player= All_Data[["match_id","player_dismissed"]]
No_Matches_player = No_Matches_player .groupby("player_dismissed")["match_id"].count().reset_index().sort_values(by="match_id",ascending=False).reset_index(drop=True)
No_Matches_player.columns=["batsman","No_of Matches"]
No_Matches_player .head(5)
```
